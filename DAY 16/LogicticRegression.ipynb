# Logistic Regression --Pass/Fail ----Men/Women
# Predict only 2 option

```python
# Step 1: import all required library
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
# Step 2: collect the data
df=pd.read_csv('pass_fail.csv')
df.head()
```

```python
# Step 3: Split the data
X = df[['Hours']]
y = df['Result']
```

```python
# Step 4: Create the model (LogisticRegression)
model= LogisticRegression()
```

```python
model.fit(X,y)
```

```python
pred_35 = model.predict([[3.5]])
print('Predicted result for 3.5 hours:', pred_35[0])
```

```python
hours_study=float(input('Number of hours studied: '))
pred_value=model.predict([[hours_study]])
print(f'After {hours_study} hours study you are likely to {pred_value[0]}')
```

    # have three variable and predict the third one

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('customer_purchase.csv')
df.head()
```

```python
# data split and Labelling
le =LabelEncoder()
df['Purchase'] = le.fit_transform(df['Purchased'])
X= df[['Age','Salary']]
y= df['Purchased']
df
```

```python
# creating the model
model=LogisticRegression()
```

```python
# trained the model
model.fit(X,y)
```

```python
# Predict the Result
new_customer=[[28,35000]]
prediction = model.predict(new_customer)
print(f"Predicted Result (1=Yes, 0=No): {prediction[0]}")
```

```python
new_cust_age=int(input('Enter Age: '))
new_cust_sal=int(input('Enter Salary: '))
new_cust =[[new_cust_age, new_cust_sal]]
prediction = model.predict(new_cust)
if (prediction):
    print('Likely Purchased')
else:
    print ('Likely Not Purchased')
```

    # Download spam_data.csv from https://www.kaggle.com/

    # Use Logistic Regression Algorithm to trained the Model
    # Predict Result if the email is spam or not spam

```python
# Step 1: import Libaries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer #Tukar data teks kepada data nombor
```

```python
# Step 2: Read file
df=pd.read_csv('spam_dataset.csv')
df
```

```python
# Step 3
X = df[['message_content']]
y = df['is_spam']
```

```python
# Step 4 (Extra Step)
cv = CountVectorizer()
X_data = cv.fit_transform(X)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_data, y)
```

```python
sample = ["Hello Cristy", "Are you coming to class today?"]
pred = model.predict(cv.transform(sample))

for msg, label in zip (sample,pred):
    print(msg, "->", "SPAM" if label == 1 else "NOT SPAM")
```
