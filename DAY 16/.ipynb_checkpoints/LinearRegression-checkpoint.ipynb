# Supervised Machine Learning
    ## Linear Regression

```python
# 1. Wajib import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# 2. Baca data dari fail csv
data=pd.read_csv('sales_data.csv')
data.head()
```

```python
# 3. Prepare data
X= data[['Advertising_Spend']] #guna double bracket sebab kita kene buat 2D array guna rows & column
y= data['Sales_Revenue'] #output
```

```python
# 4. Model Regression & fit
model = LinearRegression()
model.fit(X,y)
```

```python
# 5. Predict the Linear Regression
new_spend =float(input('Enter New Spend to Predict Sales Revenue: '))
predicted_revenue = model.predict([[new_spend]])
print(f'New Spend {new_spend} Predicted Revenue {predicted_revenue}')
```

```python
X= data[['Size_sqft']] #guna double bracket sebab kita kene buat 2D array guna rows & column
y= data['Price'] #output
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```
