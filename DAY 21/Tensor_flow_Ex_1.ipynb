```python
import numpy as np
numpy_Array=np.arange(1,25,dtype=np.int16)
numpy_Array
```

```python
tensor_numpy=tf.constant(numpy_Array,shape=(2,3,4))
tensor_numpy
```

```python
print("Tensor Information")
print("Shape:",tensor_numpy.shape)
print("Dimension:",tensor_numpy.ndim)
print("Rank:",tf.rank(tensor_numpy).numpy())
print("Size:",tf.size(tensor_numpy))
print("DataType of elements: ",tensor_numpy.dtype)
print("Element along axis 0:", tensor_numpy.shape[0])
print("Element along axis 1:", tensor_numpy.shape[1])
print("Element along axis 2:", tensor_numpy.shape[2])
print("Total number of element: \t",tf.size(tensor_numpy).numpy())
```

```python
tensor_1=tf.random.uniform(shape=(6,300),minval=0,maxval=1)

print("Tensor 1 Information")
print("Shape:",tensor_1.shape)
print("Rank:",tf.rank(tensor_1).numpy())
print("Size:",tf.size(tensor_1))
print("Total number of element: \t",tf.size(tensor_1).numpy())
```
