## First Example to create a line chart from our data
    # students.xlsx for fetching the data
    # Generate Line chart based on Student Name and Student Marks

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_excel("students.xlsx")
```

```python
names=df['Name']
marks=df['Marks']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(names,marks,marker='s',color='green')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5,3)) #size
plt.plot(names,marks,marker='o',color='orange') 
plt.title('Student Progress Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
print('Solid Line')
plt.figure(figsize=(5,4))
plt.plot(names,marks,marker='v', color='blue',linestyle='-')
plt.title('Student Progress Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
print('Dashed Line')
plt.plot(names,marks,marker='o', color='yellow',linestyle='--')
plt.show()
```

```python
print('Dotted Line')
plt.plot(names,marks,marker='o', color='yellow',linestyle=':')
plt.show()
```

```python
print('Dashed Dotted Line')
plt.plot(names,marks,marker='o', color='yellow',linestyle='-.')
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.plot(names,marks,marker='o', color='yellow',linestyle=':')
plt.title('Student Progress Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```
