```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('netflix_data.csv')
df.head(3)
```

```python
sns.set_theme(style='whitegrid',palette='Set2')
sns.countplot(data=df,x='type')
plt.title('Number of Movies VS Number of TV Shows')
plt.xlabel('Content type')
plt.ylabel('Count')
plt.show()
```

```python
df=df.dropna(subset=['type','genre'])
```

```python
df['type']
```

```python
df['genre']
```

```python
df['country']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df.head(10),x='type',hue='country',palette='Set2')
plt.title('Netflix content type of country')
plt.xlabel('Content type')
plt.ylabel('Count')
plt.legend(title='Country')
plt.show()
```

```python
plt.figure(figsize=(15,20))
sns.countplot(data=df,x='type',hue='country',palette='Set2')
plt.title('Netflix content type by country only Top 10')
plt.xlabel('Content type')
plt.ylabel('Count')
plt.legend(title='Country',bbox_to_anchor=(1.02,1), loc='upper left')
plt.tight_layout
plt.show()
```

# Using Netflix Dataset to visualize the distribution of release years for Movies vs Tv Shows using Seaborns's histplot()
    #use different color for Movies Vs Tv Show

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('netflix_data.csv')

plt.figure(figsize=(10, 5))
sns.histplot(data=df,x='release_year',hue='type',bins=30, palette='Set1')

plt.title('Distribution of Release Years: Movies vs TV Shows on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.tight_layout
plt.show()
```
