```python
import pandas as pd
```

```python
data={'Students':['Sara','Ali','Aya','Fara','Wani','Bety'],
      'Marks':[85,None,65,75,None,55]}
```

```python
df=pd.DataFrame(data)
print('Original Data')
df
```

```python
print('Missing Data')
print('Number of missing values are: ',df.isnull().sum())
```

```python
average_marks=df['Marks'].mean()
print(average_marks)
```

```python
df.fillna(average_marks, inplace=True)
```

```python
print('After fillna() call')
df
```

# Dropna() Example

```python
students_data={'Name':['Sara','Ali','Aya','Fara','Wani','Bety'], 'Marks':[85,None,65,75,None,55]}
df=pd.DataFrame(students_data)
print('Original Data')
df
```

```python
df=df.dropna(subset=['Marks'])
```

```python
print('Dataset after dropna()')
df
```
