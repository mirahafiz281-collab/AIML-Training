```python
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','Mar','April']
sales=[2000,3000,2300,3400]
```

# plt.plot(x,y,color='color name',linestyle='linestyle', linewidth=value, marker='marker symbol' label='label name')

```python
plt.plot(months,sales,color='hotpink', linestyle='--', linewidth=2, marker='d', label='Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales per Months')
plt.title('Monthly Sales Report')
plt.xlim(0,4)
plt.ylim(1000,4000)
plt.grid(True)
plt.legend(loc='upper left', fontsize=12)
plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')
plt.xlabel('X Lines')
plt.ylabel('Y Lines')
plt.title('Simple Lines Plot')
plt.legend(loc='upper left')
plt.show()
```

# Bar plot
    # plt.bar(x,y,color='edgecolor=' ')

```python
categories=['Electronics','Mobiles','Books','Groceries','Home Decoration']
sales=[50000,30000,90000,60000,40000]
bar_colors=['blue','grey','red','purple','pink']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.title('Sen Heng Sales Categories Report')
plt.grid(axis='y', linestyle='--', color='green', alpha=0.6)
plt.xticks(rotation=30, fontsize=12, color='magenta')
plt.yticks([20000,40000,60000,80000,100000],['20k','40k','60k','80k','1M'], fontsize=12, color='orange')
plt.legend(title='Legend',fontsize=8)
plt.tight_layout()
plt.show()
```

# scatter plot
    ## plt.scatter(x,y,s='',c='colorName',marker='markername', alpha='transp.' ,edgecolor='color name', label='label of legend')

```python
import matplotlib.pyplot as plt
```

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
plt.figure(figsize=(8,5))
plt.scatter(height_cm,weight_kg,s=100,c='green',marker='o',alpha=1,edgecolor='black',label='weight/height')
plt.title('Scatter Plote Height Vs Weight')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.xlim(140,190)
plt.ylim(40,90)
plt.xticks(rotation=45, fontsize=12, color='green')
plt.yticks([50,55,60,65,70,75,80,85],['50Kg','55kg','60Kg','65Kg','70kg','75kg','80kg','85kg'],fontsize=12, color='green')
plt.grid(color='lightblue',alpha=0.7,linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
