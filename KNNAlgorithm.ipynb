# KNN Algorithm in supervised Machine Learning
# KNeighborsClassifier of sklearn library for this one

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Data collection and splitting the data
X = np.array([
    [1, 3500],
    [2, 3200],
    [3, 3000],
    [4, 2500],
    [5, 2300],
    [6, 2000],
    [7, 2200],
    [2, 3700],
])
y = ['Unfit','Unfit','Unfit','Fit','Fit','Fit','Fit','Unfit']
```

```python
# Create the model
model = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the model
model.fit(X,y)
```

```python
# Predict
new_person = [[4.5, 7600]]  #4.5 hours exercise, 7600 calories
pred= model.predict(new_person)
print("Predicted Result:", pred[0])
```

```python
# Predict from taking the values from users
exercise_hours=float(input('Enter Number of Exercise Hours: '))
cal_intake=float(input('Enter number of calories consumed: '))
pred_fu= model.predict([[exercise_hours,cal_intake]])
print(f'Exercise Hours {exercise_hours}, Calories Intake {cal_intake} You are likely {pred_fu[0]}')
```

# Exercise 1

```python
import pandas as pd
```

```python
df=pd.read_csv('fruit_data.csv')
df
```

```python
# split the data
X =df[['Weight','Size']]
y =df['Fruit']
```

```python
# Create the model
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X,y)
```

```python
# Predict
# new_fruit = [[120, 2]] 
new_fruit=pd.DataFrame([[120,2]], columns= ['Weight', 'Size']) # untuk buang warning
pred= model.predict(new_fruit)
print("Predicted Result:", pred[0])
```

```python
# Predict from taking the values from users
fruit_weight=float(input('Enter weight of fruit in gram: '))
fruit_size=float(input('Enter size of fruit in cm: '))
our_fruit=pd.DataFrame([[fruit_weight,fruit_size]], columns= ['Weight', 'Size'])
pred_fruit= model.predict(our_fruit)
print(f'Fruit weight is: {fruit_weight}, Fruit size is: {fruit_size} The fruit is likely an: {pred_fruit[0]}')
```

# Task 2:
    Downlaod student_grade.csv from GitHub.com
    Predict a student's grade level (A, B, or C)
    based on hours studied and assignments completed
    using the K-Nearest Neighbors algorithm

```python
import pandas as pd
```

```python
df=pd.read_csv('student_grade.csv')
df
```

```python
X =df[['Hours_Studied','Assignments_Completed']]
y =df['Grade']
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X,y)
```

```python
new_grade=pd.DataFrame([[2,2]], columns= ['Hours_Studied', 'Assignments_Completed']) # untuk buang warning
pred= model.predict(new_grade)
print("Predicted Result:", pred[0])
```

```python
hours_studied=float(input('Enter Hours Studied: '))
ass_comp=float(input('Enter assignments completed: '))
our_grade=pd.DataFrame([[hours_studied,ass_comp]], columns= ['Hours_Studied', 'Assignments_Completed'])
pred_grade= model.predict(our_grade)
print(f'Hours Studied is: {hours_studied}, Assignments is: {ass_comp} The grade is likely: {pred_grade[0]}')
```
